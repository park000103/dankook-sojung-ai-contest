## 라이브러리 불러오기

```python
import pandas as pd
import numpy as np
```

## 2018년 강수량 데이터

- 시간별 강수량 데이터 불러오기

```python
df_2018 = pd.read_csv("2018.csv",encoding='cp949')
```

- 결측치 0으로 채우기

```python
df_2018["강수량(mm)"] = df_2018["강수량(mm)"].replace(np.nan, 0)
```

- 필요한 변수 추출

```python
df_2018.drop(['강수량 QC플래그','지점명','지점'], axis=1,  inplace=True)
```

- 날짜를 index로 시간을 column으로 pivot table 생성

```python
splited = df_2018['일시'].str.split(' ', expand=True)
splited.columns = ['날짜','시간']
new_2018 = pd.concat([df_2018,splited],axis=1)
new_2018 = new_2018.pivot(index='날짜', columns='시간', values='강수량(mm)')
```

```python
new_2018
```

### 활동시간(7:00~23:00)에 대한 강수량 합

```python
df2_2018 = pd.read_csv('2018_2.csv',index_col=0)
```

```python
com_2018 = df2_2018.combine_first(new_2018)
com_2018 = com_2018.replace(np.nan, 0)
active_2018 = com_2018[['7:00','8:00','9:00','10:00','11:00','12:00','13:00','14:00','15:00','16:00','17:00','18:00','19:00','20:00','21:00','22:00','23:00']]
active_2018['sum']=active_2018.sum(axis=1)
active_2018.drop(['7:00','8:00','9:00','10:00','11:00','12:00','13:00','14:00','15:00','16:00','17:00','18:00','19:00','20:00','21:00','22:00','23:00'], axis=1,  inplace=True)
active_2018 = active_2018.rename(columns={'sum':'act_sum'})
active_2018
```

### 출근시간(5:00~10:00)에 대한 강수량 합

```python
workPre_2018 = com_2018[['5:00','6:00','7:00','8:00','9:00','10:00']]
workPre_2018['sum']=workPre_2018.sum(axis=1)
workPre_2018.drop(['5:00','6:00','7:00','8:00','9:00','10:00'], axis=1,  inplace=True)
workPre_2018 = workPre_2018.rename(columns={'sum':'work_sum'})
workPre_2018
```

### 퇴근시간(16:00~22:00) 강수량에 대한 합

```python
homePre_2018 = com_2018[['16:00','17:00','18:00','19:00','20:00','21:00','22:00']]
homePre_2018['sum']=homePre_2018.sum(axis=1)
homePre_2018.drop(['16:00','17:00','18:00','19:00','20:00','21:00','22:00'], axis=1,  inplace=True)
homePre_2018 = homePre_2018.rename(columns={'sum':'home_sum'})
homePre_2018
```

### 활동/출근/퇴근시간에 따른 강수량의 합 자료 병합

```python
Pre_2018 = pd.concat([active_2018, workPre_2018], axis=1)
Pre_2018 = pd.concat([Pre_2018, homePre_2018], axis=1)
Pre_2018
```

- 자료 내보내기

```python
Pre_2018.to_csv('pre2018.csv', index = True, encoding = 'UTF-8')
```
